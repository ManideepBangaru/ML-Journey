```python
def Imputer(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        elif Pclass == 3:
            return 24
    else:
        return Age
```

```python
df['Age'] = df[['Age','Pclass']].apply(Imputer, axis = 1)
```

```python
plt.figure(figsize = (12,7))
sns.heatmap(df.isnull())
```

#### Converting Categorical features

```python
df.info()
```

```python
pd.get_dummies(df['Sex'], drop_first=True).head()
```

```python
sex = pd.get_dummies(df['Sex'], drop_first=1)
Embarked = pd.get_dummies(df['Embarked'], drop_first=1)
```

```python
df.drop(['Sex','Embarked','Name','Ticket'], axis = 1,inplace=True)
```

```python
df.head()
```

```python
df = pd.concat([df,sex,Embarked], axis = 1)
```

```python
df.head()
```

### Building a logistic regression model

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(df.drop('Survived', axis = 1),
                                                   df['Survived'],
                                                   test_size = 0.30,
                                                   random_state = 123)
```

#### Training and Predicting

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.neighbors import KNeighborsClassifier
LogModel = LogisticRegression()
```

```python
pipe1 = make_pipeline(StandardScaler(), LogisticRegression())
pipe1.fit(X_train, Y_train)
```

```python
pipe1.score(X_test, Y_test)
```

```python
predictions1 = pipe1.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
accuracy1 = confusion_matrix(Y_test, predictions1)
```

```python
accuracy1
```

```python
pipe2 = make_pipeline(StandardScaler(),KNeighborsClassifier(n_neighbors=10))
pipe2.fit(X_train, Y_train)
```

```python
pipe2.score(X_test,Y_test)
```

```python
predictions2 = pipe2.predict(X_test)
```

```python
accuracy2 = confusion_matrix(Y_test,predictions2)
```

```python
accuracy2
```

```python
from sklearn.metrics import accuracy_score
accuracyAuut1 = accuracy_score(Y_test,predictions1)
print('Accuracy with Logistic regression is %s'%(round(accuracyAuut1,2)))
accuracyAuut2 = accuracy_score(Y_test,predictions2)
print('Accuracy with KNN Classifier is %s'%(round(accuracyAuut2,2)))
```
